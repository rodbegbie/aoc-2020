```python
starting_numbers = (1,0,16,5,17,4)

age = 0
last_num = -1
last_seen = {}

for num in starting_numbers:
    age += 1
    last_seen[last_num] = age
    last_num = num
    # print (age, num, last_seen)

while age < 30000000:
    age += 1
    num = age-last_seen[last_num] if last_num in last_seen else 0
    if not age % 10000:
        print (age, num)
    last_seen[last_num] = age
    last_num = num

print ("** ANSWER **", age, num)
```
