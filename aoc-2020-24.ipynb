```python
flipped = set()

for direction in input_str.splitlines():
    cur_tile = (0,0,0)
    while direction:
        d = direction[0]
        direction = direction[1:]
        if d in ("n", "s"):
            d += direction[0]
            direction = direction[1:]
        # print ("AT", cur_tile, "GOING", d)

        (a,r,c) = cur_tile
        if d == "e":
            cur_tile = (a, r, c + 1)
        elif d == "w":
            cur_tile = (a, r, c - 1)
        elif d == "ne":
            cur_tile = (1-a, r-(1-a), c+a)
        elif d == "nw":
            cur_tile = (1-a, r-(1-a), c-(1-a))
        elif d == "se":
            cur_tile = (1-a, r+a, c+a)
        elif d == "sw":
            cur_tile = (1-a, r+a, c-(1-a))
        

    print ("FLIPPING", cur_tile, "TO", "WHITE" if cur_tile in flipped else "BLACK")
    if cur_tile in flipped:
        flipped.remove(cur_tile)
    else:
        flipped.add(cur_tile)

print ("TOTAL FLIPPED", len(flipped))
```

```python
print (f"Day 0: {len(flipped)}")

for day in range(1, 101):
    min_a = min([a for (a,r,c) in flipped])
    max_a = max([a for (a,r,c) in flipped])
    min_r = min([r for (a,r,c) in flipped])
    max_r = max([r for (a,r,c) in flipped])
    min_c = min([c for (a,r,c) in flipped])
    max_c = max([c for (a,r,c) in flipped])

    new_flipped = set(flipped)

    for a in range(min_a - 1, max_a + 2):
        for r in range(min_r - 1, max_r + 2):
            for c in range(min_c - 1, max_c + 2):
                count = 0
                for t in (
                    (a, r, c + 1),
                    (a, r, c - 1),
                    (1-a, r-(1-a), c+a),
                    (1-a, r-(1-a), c-(1-a)),
                    (1-a, r+a, c+a),
                    (1-a, r+a, c-(1-a)),
                ):
                    if t in flipped:
                        count += 1
                    
                if (a,r,c) in flipped and (count == 0 or count > 2):
                    new_flipped.remove((a,r,c))
                elif (a,r,c) not in flipped and count == 2:
                    new_flipped.add((a,r,c))

    flipped = new_flipped
    print (f"Day {day}: {len(flipped)}")

```
