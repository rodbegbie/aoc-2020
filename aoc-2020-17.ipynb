```python
input_str = """.#.#.#..
..#....#
#####..#
#####..#
#####..#
###..#.#
#..##.##
#.#.####"""

# input_str = """.#.
# ..#
# ###"""
```

```python
from collections import defaultdict
from copy import deepcopy

grid = defaultdict(lambda: defaultdict(lambda: defaultdict(lambda: False)))

range_xy = len(input_str.splitlines()) // 2
range_z = 0

row = -range_xy
for row_str in input_str.splitlines():
    col = -range_xy
    for char in row_str:
        if char == "#":
            grid[0][row][col] = True
        col += 1
    row += 1

def print_grid():
    lit = 0
    for z in range(-range_z, range_z + 1):
        print ("\nz=", z)
        for x in range(-range_xy, range_xy + 1):
            chars=[]
            for y in range(-range_xy, range_xy + 1):
                chars.append("#" if grid[z][x][y] else ".")
                lit += 1 if grid[z][x][y] else 0
            print("".join(chars))

    print ("NUMBER LIT", lit)

print_grid()

for i in range(6):
    old_grid = deepcopy(grid)
    range_z += 1
    range_xy += 1

    for z in range(-range_z, range_z + 1):
        for x in range(-range_xy, range_xy + 1):
            for y in range(-range_xy, range_xy + 1):
                currently_lit = old_grid[z][x][y]

                num_lit = 0
                for check_z in range(z-1, z+2):
                    for check_x in range(x-1, x+2):
                        for check_y in range(y-1, y+2):
                            if check_z == z and check_y == y and check_x == x:
                                continue
                            num_lit += 1 if old_grid[check_z][check_x][check_y] else 0
                
                if currently_lit and num_lit not in (2, 3):
                    grid[z][x][y] = False
                elif not currently_lit and num_lit == 3:
                    grid[z][x][y] = True

    print ("\n\n ***** ITERATION", i)
    print_grid()
```
