```python
tiles = {}

for tile_str in input_str.split("\n\n"):
    tile_lines = tile_str.splitlines()
    tile_id = tile_lines[0][5:-1]
    sides = set([
        tile_lines[1],
        tile_lines[10],
        "".join(line[0] for line in tile_lines[1:]),
        "".join(line[-1] for line in tile_lines[1:]),
    ])
    for side in list(sides):
        sides.add(side[::-1])
    tiles[tile_id] = sides
    # print ("TILE_ID", tile_id, "SIDES", sides)

corners = 1
corner_piece_id = ""
for (tile_id, sides) in tiles.items():
    matches = 0
    for (cmp_tile_id, cmp_sides) in tiles.items():
        if tile_id == cmp_tile_id:
            continue
        if sides & cmp_sides:
            # print ("TILE", tile_id, "MATCHES", cmp_tile_id)
            matches += 1
    if matches == 2:
        print ("**** CORNER PIECE", tile_id)
        corner_piece_id = tile_id
        corners *= int(tile_id)

print ("CORNERS", corners)
print ("CORNER_PIECE_ID FOR PART TWO!", corner_piece_id)
    
```

```python
from collections import defaultdict

tops = defaultdict(set)
lefts = defaultdict(set)

def tlbr(tile):
    top = tile.splitlines()[0]
    bottom = tile.splitlines()[-1]
    left = "".join(line[0] for line in tile.splitlines())
    right = "".join(line[-1] for line in tile.splitlines())
    return (top, left, bottom, right)

def flip_lines_x(tile_lines):
    retval = []
    for line in tile_lines:
        retval.append(line[::-1])
    return retval

def rotate_lines(tile_lines):
    retval = []
    for i in range(len(tile_lines)):
        retval.append("".join([line[i] for line in tile_lines[::-1]]))
    return retval

def tile_permutations(tile):
    perms = []
    tile_lines = tile.splitlines()
    for flip_x in (0, 1):
        for rotate in (0, 1, 2, 3):
            # print (flip_x, rotate)
            tile = tile_lines
            if flip_x:
                tile = flip_lines_x(tile)
            # if flip_y:
                # tile = flip_lines_y(tile)
            for i in range(rotate):
                tile = rotate_lines(tile)
            perms.append(("\n".join(tile)))
            # print ("\n".join(tile))
    return perms

print (tile_permutations("123\n456\n789"))

corner_piece = []
print ("BUILDING TILE DICTS")
for tile_str in input_str.split("\n\n"):
    tile_id = tile_str.splitlines()[0][5:-1]
    tile_lines = "\n".join(tile_str.splitlines()[1:])
    for tile in tile_permutations(tile_lines):
        if tile_id == corner_piece_id:
            corner_piece = tile
        (top, left, bottom, right) = tlbr(tile)
        tops[top].add(tile)
        lefts[left].add(tile)

print ("BUILDING PHOTO")
picture_tiles = []
current_row = []

def remove_tile(tile):
    for perm in tile_permutations(tile):
        (top, left, bottom, right) = tlbr(perm)
        tops[top].remove(perm)
        lefts[left].remove(perm)

remove_tile(corner_piece)
for tile in tile_permutations(corner_piece):
    (top, left, bottom, right) = tlbr(tile)
    # print ("\nCONSIDERING")
    # print (tile)
    if (tops[bottom] and lefts[right]):
        print ("FIRST_TILE!")
        print (tile)
        current_row.append(tile)
        break

while True:
    while True:
        prev_right = tlbr(current_row[-1])[3]
        if not lefts[prev_right]:
            break
        print ("MOVING RIGHT", prev_right)
        next_tile = list(lefts[prev_right])[0]
        current_row.append(next_tile)
        remove_tile(next_tile)
    print ("MOVING DOWN")
    picture_tiles.append(current_row)
    prev_bottom = current_row[0].splitlines()[-1]
    print ("PREV_BOT", prev_bottom)
    if not tops[prev_bottom]:
        break
    next_tile = list(tops[prev_bottom])[0]
    current_row = [next_tile]
    remove_tile(next_tile)

picture = ""
for tile_row in picture_tiles:
    rows = [""] * 8
    for i in range(8):
        for tile in tile_row:
            rows[i] += tile.splitlines()[i+1][1:-1]
    for row in rows:
        picture += row + "\n"

print ("PICTURE")
print (picture)

print ("MONSTER HUNTING!")

monster = """                  # 
#    ##    ##    ###
 #  #  #  #  #  #   """

monster_width = len(monster.splitlines()[0])
monster_height = len(monster.splitlines())

def count_hashes(string):
    return sum([(1 if c == "#" else 0) for c in string])

monster_hashes = count_hashes(monster)

def find_monster(string):
    for pos in range(len(monster)):
        if monster[pos] == "#" and string[pos] != "#":
            return False
    return True

for perm in tile_permutations(picture):
    print ("===> ROTATION")
    print (perm)
    
    found = 0
    for x in range(len(picture.splitlines()[0]) - monster_width):
        for y in range(len(picture.splitlines()) - monster_height):
            # print ("X", x, "Y", y)
            area = "\n".join([
                perm.splitlines()[y][x:x+monster_width],
                perm.splitlines()[y+1][x:x+monster_width],
                perm.splitlines()[y+2][x:x+monster_width]
            ])
            # print ("AREAAA")
            # print (area)
            if find_monster(area):
                print ("FOUND!")
                found += 1
    if found:
        print ("MONSTER HASHES", found * monster_hashes)
        print ("OTHER HASHES", count_hashes(picture) - (found * monster_hashes))
        break
```
