```python
(p1_str, p2_str) = input_str.split("\n\n")
p1 = [int(x) for x in p1_str.splitlines()[1:]]
p2 = [int(x) for x in p2_str.splitlines()[1:]]

while p1 and p2:
    c1 = p1.pop(0)
    c2 = p2.pop(0)
    if c1 > c2:
        p1.append(c1)
        p1.append(c2)
    else:
        p2.append(c2)
        p2.append(c1)

winner = p1 if p1 else p2

score = 0
for i in range(1, len(winner)+1):
    score += i * winner[-i]

score
```

```python
(p1_str, p2_str) = input_str.split("\n\n")
p1 = [int(x) for x in p1_str.splitlines()[1:]]
p2 = [int(x) for x in p2_str.splitlines()[1:]]

game_counter = 0

def game(p1, p2):
    global game_counter
    game_counter += 1

    game_num = game_counter
    prev_p1 = []
    prev_p2 = []

    # print (f"=== Game {game_num} ===\n")

    round_num = 0
    while True:
        round_num += 1

        # print (f"-- Round {round_num} (Game {game_num})")
        # print ("Player 1's deck:", " ".join([str(x) for x in p1]))
        # print ("Player 2's deck:", " ".join([str(x) for x in p2]))

        winner = 0
        if p1 in prev_p1 and p2 in prev_p2:
            print (f"The winner of game {game_num} is player 1!")
            return (1, p1)

        prev_p1.append(list(p1))
        prev_p2.append(list(p2))
        
        c1 = p1.pop(0)
        c2 = p2.pop(0)
        # print (f"Player 1 plays: {c1}")
        # print (f"Player 2 plays: {c2}")

        if len(p1) >= c1 and len(p2) >= c2:
            # print ("Playing a sub-game to determine the winner...\n")
            (winner, _) = game(p1[:c1], p2[:c2])
        else:
            if c1 > c2:
                winner = 1
            else:
                winner = 2
        
        if winner == 1:
            # print (f"Player 1 wins round {round_num} of game {game_num}!\n")
            p1.append(c1)
            p1.append(c2)
        else:
            # print (f"Player 2 wins round {round_num} of game {game_num}!\n")
            p2.append(c2)
            p2.append(c1)
        
        if not p1:
            print (f"The winner of game {game_num} is player 2!")
            return (2, p2)
        elif not p2:
            print (f"The winner of game {game_num} is player 1!")
            return (1, p1)

(_, winner) = game(p1, p2)

score = 0
for i in range(1, len(winner)+1):
    score += i * winner[-i]

score
```
