```python
current = 0
differences = {1: 0, 2: 0, 3: 0}

for adapter in adapters[1:]:
    diff = adapter-current
    differences[diff] += 1
    current = adapter

print (differences[1], "*", differences[3], "=", differences[1] * differences[3])
```

```python
pos = len(adapters) - 2
paths_from_here = {adapters[-1]: 1}
while pos >= 0:
    adapter = adapters[pos]
    steps = 0
    for i in (1,2,3):
        if (adapter + i) in adapters:
            steps += paths_from_here[adapter + i]
    
    paths_from_here[adapter] = steps

    # print ("adapter", adapter, "steps", steps, "paths_from_here", paths_from_here)
    pos -= 1

print ("PATHS", paths_from_here[adapters[0]])
```
